```python
import matplotlib.pyplot as plt
import time
import psutil
```

```python
%matplotlib notebook

plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
psutil.cpu_percent()
```

```python
i = 0
x, y = [], []
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y)
    fig.canvas.draw()
    time.sleep(0.1)
    i = i+1
```
